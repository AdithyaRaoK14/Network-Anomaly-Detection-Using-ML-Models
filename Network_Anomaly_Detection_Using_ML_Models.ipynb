# Binary Classification

```python
# =========================
# Step 0: Imports
# =========================
import os
import pandas as pd
import joblib
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
from sklearn.metrics import (
    confusion_matrix, classification_report, roc_auc_score, roc_curve,
    precision_score, recall_score, f1_score
)

# =========================
# Step 1: Load CSVs
# =========================
data_path = './data/cicids2017'
files = [f for f in os.listdir(data_path) if f.endswith('.csv')]
print("Files found:", files)

df_list = [pd.read_csv(os.path.join(data_path, f)) for f in files]
df = pd.concat(df_list, ignore_index=True)
print("Dataset shape:", df.shape)

# =========================
# Step 2: Binary classification
# =========================
df['Binary_Label'] = (df['Attack Type'] != 'Normal Traffic').astype(int)

X = df.drop(['Attack Type', 'Binary_Label'], axis=1)
y = df['Binary_Label']

# =========================
# Step 2.1: Feature Selection
# =========================
selector = SelectKBest(score_func=f_classif, k=30)
X_selected = selector.fit_transform(X, y)
selected_features = X.columns[selector.get_support()]
print("Selected Features:", list(selected_features))

# =========================
# Step 3: Scale Features
# =========================
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_selected)

# =========================
# Step 4: Train-test split
# =========================
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)

# =========================
# Step 5: Function to save model & outputs
# =========================
def save_model_outputs(model, model_name, X_test, y_test, base_folder='./models'):
    folder_path = os.path.join(base_folder, model_name)
    os.makedirs(folder_path, exist_ok=True)

    # Save model
    joblib.dump(model, os.path.join(folder_path, f'{model_name}.joblib'))

    # Predictions
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:,1]

    # Confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    print(f"\n=== {model_name} Confusion Matrix ===")
    print(cm)
    plt.figure(figsize=(6,5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Normal','Attack'], yticklabels=['Normal','Attack'])
    plt.title(f'{model_name} Confusion Matrix')
    plt.savefig(os.path.join(folder_path, f'{model_name}_confusion_matrix.png'))
    plt.close()

    # Classification report
    report = classification_report(y_test, y_pred)
    print(f"\n=== {model_name} Classification Report ===")
    print(report)
    with open(os.path.join(folder_path, f'{model_name}_classification_report.txt'), 'w') as f:
        f.write(report)

    # ROC curve
    roc_auc = roc_auc_score(y_test, y_prob)
    fpr, tpr, _ = roc_curve(y_test, y_prob)
    plt.figure()
    plt.plot(fpr, tpr, label=f'ROC curve (AUC = {roc_auc:.4f})')
    plt.plot([0, 1], [0, 1], 'k--')
    plt.title(f'{model_name} ROC Curve')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.legend(loc='lower right')
    plt.savefig(os.path.join(folder_path, f'{model_name}_roc_curve.png'))
    plt.close()

    # Save ROC-AUC
    with open(os.path.join(folder_path, f'{model_name}_roc_auc.txt'), 'w') as f:
        f.write(f'ROC-AUC: {roc_auc}\n')

    # Return metrics
    return {
        'Model': model_name,
        'ROC-AUC': roc_auc,
        'Precision': precision_score(y_test, y_pred),
        'Recall': recall_score(y_test, y_pred),
        'F1-score': f1_score(y_test, y_pred)
    }

# =========================
# BLOCK 1: RandomForest + DecisionTree
# =========================
results_main = []

# 1ï¸âƒ£ Random Forest
rf = RandomForestClassifier(n_estimators=100, class_weight='balanced', random_state=42)
rf.fit(X_train, y_train)
results_main.append(save_model_outputs(rf, 'RandomForest', X_test, y_test))

# 2ï¸âƒ£ Decision Tree
dt = DecisionTreeClassifier(class_weight='balanced', random_state=42)
dt.fit(X_train, y_train)
results_main.append(save_model_outputs(dt, 'DecisionTree', X_test, y_test))

summary_df_main = pd.DataFrame(results_main)
summary_df_main.to_csv('./models/models_summary_main.csv', index=False)
print("\nâœ… RandomForest & DecisionTree models trained and saved!")
print(summary_df_main)
```

```python
# =========================
# BLOCK 2: XGBoost + AdaBoost
# =========================
results_boost = []

# 3ï¸âƒ£ XGBoost
xgb = XGBClassifier(
    n_estimators=100,
    scale_pos_weight=(y_train==0).sum()/(y_train==1).sum(),
    use_label_encoder=False,
    eval_metric='logloss',
    random_state=42
)
xgb.fit(X_train, y_train)
results_boost.append(save_model_outputs(xgb, 'XGBoost', X_test, y_test))

# 4ï¸âƒ£ AdaBoost
ada = AdaBoostClassifier(n_estimators=100, random_state=42)
ada.fit(X_train, y_train)
results_boost.append(save_model_outputs(ada, 'AdaBoost', X_test, y_test))

summary_df_boost = pd.DataFrame(results_boost)
summary_df_boost.to_csv('./models/models_summary_boost.csv', index=False)
print("\nâœ… XGBoost & AdaBoost models trained and saved!")
print(summary_df_boost)
```

```python
import gradio as gr
import numpy as np
import joblib

# =========================
# Load models, scaler, selector
# =========================
models = {
    "Random Forest": joblib.load('./models/RandomForest/RandomForest.joblib'),
    "Decision Tree": joblib.load('./models/DecisionTree/DecisionTree.joblib'),
    "XGBoost": joblib.load('./models/XGBoost/XGBoost.joblib'),
    "AdaBoost": joblib.load('./models/AdaBoost/AdaBoost.joblib')
}

scaler = joblib.load('./models/scaler.joblib')        # saved StandardScaler
selector = joblib.load('./models/selector.joblib')    # saved SelectKBest

# Binary class labels
classes = ['Normal Traffic', 'Attack']

# =========================
# Prediction function
# =========================
def predict_network_flow(selected_model, *features):
    """
    selected_model: string, one of the models in models dict
    features: float values of all selected features
    """
    X_input = np.array(features).reshape(1, -1)

    # Apply feature selection & scaling
    X_selected = selector.transform(X_input)
    X_scaled = scaler.transform(X_selected)

    model = models[selected_model]

    # Predict class
    y_pred = model.predict(X_scaled)[0]

    # Predict probability (if available)
    if hasattr(model, "predict_proba"):
        y_prob = model.predict_proba(X_scaled)[0][1]
        return f"{classes[y_pred]} (Probability of Attack: {y_prob:.2f})"
    else:
        return f"{classes[y_pred]}"

# =========================
# Gradio interface
# =========================
# Number of selected features
num_features = selector.transform(np.zeros((1, selector.n_features_in_))).shape[1]
input_labels = [f"Feature {i+1}" for i in range(num_features)]

inputs = [
    gr.Dropdown(choices=list(models.keys()), label="Select Model")
] + [gr.Number(label=lbl) for lbl in input_labels]

iface = gr.Interface(
    fn=predict_network_flow,
    inputs=inputs,
    outputs="text",
    title="Binary Network Anomaly Detection",
    description="Choose a model and enter network flow features to predict Normal Traffic or Attack"
)

iface.launch()
```

```python
# =========================
# Step 0: Imports
# =========================
import os
import pandas as pd
import joblib
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
from sklearn.metrics import confusion_matrix, classification_report, precision_score, recall_score, f1_score, roc_auc_score, roc_curve

# =========================
# Step 1: Load all CSV files
# =========================
data_path = './data/cicids2017'   # Your dataset folder
files = [f for f in os.listdir(data_path) if f.endswith('.csv')]
print("Files found:", files)

df_list = [pd.read_csv(os.path.join(data_path, f)) for f in files]
df = pd.concat(df_list, ignore_index=True)
print("Dataset shape:", df.shape)

# =========================
# Step 2: Prepare multi-class labels
# =========================
X = df.drop(['Attack Type'], axis=1)  # remove only 'Attack Type'
y = df['Attack Type']

# Encode labels as integers
le = LabelEncoder()
y_encoded = le.fit_transform(y)
classes = le.classes_
print("Classes:", classes)


# =========================
# Step 2.1: Feature Selection
# =========================
selector = SelectKBest(score_func=f_classif, k=30)
X_selected = selector.fit_transform(X, y_encoded)
selected_features = X.columns[selector.get_support()]
print("Selected Features:", list(selected_features))

# =========================
# Step 3: Scale Features
# =========================
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_selected)

# =========================
# Step 4: Train-test split
# =========================
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)

# =========================
# Step 5: Function to save model & outputs & print
# =========================
def save_model_outputs(model, model_name, X_test, y_test, label_encoder, base_folder='./models_multiclass'):
    folder_path = os.path.join(base_folder, model_name)
    os.makedirs(folder_path, exist_ok=True)

    # Save model
    joblib.dump(model, os.path.join(folder_path, f'{model_name}.joblib'))

    # Predictions
    y_pred = model.predict(X_test)

    # Confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    print(f"\n=== {model_name} Confusion Matrix ===")
    print(cm)
    plt.figure(figsize=(10,8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
    plt.title(f'{model_name} Confusion Matrix')
    plt.show()
    plt.savefig(os.path.join(folder_path, f'{model_name}_confusion_matrix.png'))
    plt.close()

    # Classification report
    report = classification_report(y_test, y_pred, target_names=label_encoder.classes_)
    print(f"\n=== {model_name} Classification Report ===")
    print(report)
    with open(os.path.join(folder_path, f'{model_name}_classification_report.txt'), 'w') as f:
        f.write(report)

    # ROC-AUC per class
    try:
        y_prob = model.predict_proba(X_test)
        roc_auc_dict = {}
        for i, class_name in enumerate(label_encoder.classes_):
            roc_auc = roc_auc_score((y_test==i).astype(int), y_prob[:,i])
            roc_auc_dict[class_name] = roc_auc
            fpr, tpr, _ = roc_curve((y_test==i).astype(int), y_prob[:,i])
            plt.plot(fpr, tpr, label=f'{class_name} (area = {roc_auc:.4f})')
        plt.plot([0, 1], [0, 1], 'k--')
        plt.title(f'{model_name} Multi-class ROC Curve')
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.legend(loc='lower right')
        plt.show()
        plt.savefig(os.path.join(folder_path, f'{model_name}_roc_curve.png'))
        plt.close()

        with open(os.path.join(folder_path, f'{model_name}_roc_auc.txt'), 'w') as f:
            for class_name, auc in roc_auc_dict.items():
                f.write(f'{class_name}: {auc}\n')
    except:
        print(f"ROC-AUC could not be computed for {model_name} (model may not support predict_proba).")

    precision = precision_score(y_test, y_pred, average='weighted')
    recall = recall_score(y_test, y_pred, average='weighted')
    f1 = f1_score(y_test, y_pred, average='weighted')

    return {'Model': model_name, 'Precision': precision, 'Recall': recall, 'F1-score': f1}

# =========================
# Step 6: Train models & save outputs
# =========================
results = []

# 6.1 Random Forest
rf = RandomForestClassifier(n_estimators=100, class_weight='balanced', random_state=42)
rf.fit(X_train, y_train)
results.append(save_model_outputs(rf, 'RandomForest', X_test, y_test, le))

# 6.2 Decision Tree
dt = DecisionTreeClassifier(class_weight='balanced', random_state=42)
dt.fit(X_train, y_train)
results.append(save_model_outputs(dt, 'DecisionTree', X_test, y_test, le))

# 6.3 XGBoost
xgb = XGBClassifier(
    n_estimators=100,
    use_label_encoder=False,
    eval_metric='mlogloss',
    random_state=42
)
xgb.fit(X_train, y_train)
results.append(save_model_outputs(xgb, 'XGBoost', X_test, y_test, le))

# 6.4 AdaBoost
ada = AdaBoostClassifier(n_estimators=100, random_state=42)
ada.fit(X_train, y_train)
results.append(save_model_outputs(ada, 'AdaBoost', X_test, y_test, le))

# =========================
# Step 7: Save summary CSV
# =========================
summary_df = pd.DataFrame(results)
summary_df.to_csv('./models_multiclass/models_summary.csv', index=False)
print("\nAll multi-class models trained and summary saved!")
print(summary_df)
```
